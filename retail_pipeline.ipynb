```python
retail_df['transaction_date'] = pd.to_datetime(
    retail_df['transaction_date'],
    errors='coerce'
)

retail_df['year'] = retail_df['transaction_date'].dt.year
retail_df['month'] = retail_df['transaction_date'].dt.month_name()
retail_df['quarter'] = retail_df['transaction_date'].dt.quarter
```

```python
product_df.columns
```

```python
retail_df = retail_df.merge(
    product_df,
    on='product_id',
    how='left',
    suffixes=('', '_master')
)

print(retail_df.shape)
```

```python
retail_df.columns
```

```python
retail_df['price'] = retail_df['price'].fillna(
    retail_df['price_master']
)
```

```python
retail_df['product_name'] = retail_df['product_name_master']

retail_df['category'] = retail_df['category_master']
```

```python
def mask_email(email):
    try:
        parts = str(email).split("@")
        return parts[0][:2] + "****@" + parts[1]
    except:
        return email

retail_df['masked_email'] = retail_df['email'].apply(mask_email)
```

```python
def mask_phone(phone):
    phone = str(phone)
    return "******" + phone[-4:]

retail_df['masked_phone'] = retail_df['phone'].apply(mask_phone)
```

```python
retail_df['revenue'] = (
    retail_df['quantity']
    * retail_df['price']
    * (1 - retail_df['discount'])
)
```

```python
revenue_by_category = retail_df.groupby(
    'category'
)['revenue'].sum().reset_index()

revenue_by_category
```

```python
revenue_by_city = retail_df.groupby(
    'city'
)['revenue'].sum().reset_index()

revenue_by_city
```

```python
monthly_revenue = retail_df.groupby(
    ['year', 'month']
)['revenue'].sum().reset_index()

monthly_revenue
```

```python
retail_df.to_csv(
    "../Data/Processed/cleaned_retail_sales.csv",
    index=False
)
```
